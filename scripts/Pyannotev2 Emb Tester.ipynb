```python
import pandas as pd
from pyannote.audio import Inference
import numpy as np

toy = '/Users/rahulbrito/Documents/projects/infantvoice/data/Full_Readings/0300222_recordings_16khz/020.wav'
#toy2 ='/Users/rahulbrito/Documents/projects/infantvoice/data/Full_Readings/0300222_recordings_16khz/002.wav'
```

```python
inference = Inference("pyannote/embedding", 
                      window="sliding",
                      duration=3.0, step=1.0)
embeddings = inference(toy)
# `embeddings` is a (N x D) pyannote.core.SlidingWindowFeature
# `embeddings[i]` is the embedding of the ith position of the 
# sliding window, i.e. from [i * step, i * step + duration].
```

```python
%load_ext autoreload
%autoreload 2
import os
from pyannote.core import json
import src.embedding_extractor as ee
import pandas as pd
from pathlib import Path

diarized_dir = '/Users/rahulbrito/Documents/projects/infantvoice/data/Full_Readings/031622_diarized'
diarized = [(json.load_from(os.path.join(diarized_dir,file)),file) for file in os.listdir(diarized_dir)]
data_clean_lower = {int(Path(part).stem):data.label_timeline('SPEAKER_01') for data,part in diarized if int(Path(part).stem)<11}
data_clean_upper = {int(Path(part).stem):data.label_timeline('SPEAKER_00') for data,part in diarized if int(Path(part).stem)>=11}
data_clean = data_clean_lower|data_clean_upper
```

```python
emb_from_sample = []
for segment in data_clean[20]:
        inter = embeddings.crop(segment, 'strict')
        emb_from_sample.append(inter)
emb_from_sample = pd.DataFrame(np.vstack(emb_from_sample))
```
